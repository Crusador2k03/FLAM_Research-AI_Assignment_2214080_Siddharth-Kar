```python
import numpy as np
from io import StringIO
from math import degrees
import sys

try:
    from scipy.optimize import minimize, minimize_scalar
    SCIPY = True
except Exception:
    SCIPY = False
    print("Warning: scipy not available. The script will use a fallback grid search.", file=sys.stderr)

data_str = r"""
88.364456,57.784378
74.283936,54.40678
60.256474,46.311462
82.13437,57.717567
101.03639,67.84934
69.284134,48.406086
77.81013,57.588833
91.29688,63.03135
107.66841,65.91405
76.40689,56.871525
72.313515,51.168285
100.56871,68.07376
85.40742,57.31956
65.73637,47.454475
78.559,57.776672
102.01643,67.376236
92.44081,67.234024
102.319664,67.23141
73.051186,52.32614
70.23932,48.974834
88.56056,57.891556
91.96877,65.819046
86.83264,57.347816
92.62934,67.65634
104.750336,66.19454
73.5264,53.136063
93.27278,68.66579
71.30269,49.905098
64.975815,47.35335
89.05592,58.24083
100.71979,68.00157
106.31808,65.80417
108.06331,66.10597
67.84049,47.865753
82.02788,57.731422
98.86425,68.84363
72.67582,51.715893
97.88674,69.218376
88.70565,57.981354
70.66043,49.300117
95.17704,69.66583
90.98131,61.803932
74.11193,54.128273
67.423485,47.758686
75.07062,55.54406
93.32456,68.72479
92.49515,67.36309
109.09305,67.57011
74.6767,55.005707
84.74091,57.372154
83.09207,57.584
78.128586,57.682087
65.000305,47.356487
92.592575,67.57944
94.47159,69.50913
84.89339,57.357613
92.58727,67.568146
89.3978,58.564568
85.80336,57.30451
109.05394,67.45857
72.61481,51.620525
84.3335,57.416786
91.34593,63.23906
91.305405,63.06719
62.731583,47.02679
84.717255,57.374527
108.320114,66.29982
93.38754,68.792984
87.94448,57.602074
74.52983,54.788445
98.66312,68.92605
73.594505,53.253242
84.93126,57.354206
67.8154,47.858814
73.32133,52.783768
70.60834,49.25702
84.30343,57.420364
89.181885,58.35109
71.52109,50.144382
102.03093,67.36928
92.67618,67.75073
96.70082,69.55569
63.39724,47.13958
73.5922,53.24928
87.72062,57.527298
91.83847,65.328674
100.10727,68.291534
72.33709,51.20249
71.38096,49.98879
79.996185,57.89074
66.2157,47.525764
94.11699,69.35312
106.67454,65.77663
65.80095,47.463623
104.915665,66.138336
68.09816,47.941082
76.40583,56.87075
97.43371,69.365585
91.13893,62.39004
64.61357,47.307106
75.18191,55.68387
79.86053,57.89057
106.91785,65.777115
80.97552,57.84513
75.69756,56.260036
104.00347,66.47797
85.33286,57.3239
62.00487,46.874073
60.975353,46.582985
64.781975,47.3286
68.08688,47.937626
67.83224,47.863464
105.982315,65.85555
73.884186,53.748222
86.084404,57.303146
74.4463,54.66118
105.78893,65.89461
107.52535,65.8685
73.66033,53.366364
64.37273,47.276096
73.5063,53.101463
63.259483,47.117928
94.03549,69.30809
68.93807,48.247204
91.60011,64.34067
78.77941,57.81267
75.13112,55.620747
66.16168,47.517296
65.39946,47.408455
87.790306,57.549076
72.77513,51.87362
91.200226,62.633213
102.7788,67.01567
76.0424,56.58198
75.27876,55.80104
87.3976,57.442387
102.9538,66.93498
87.94472,57.60216
74.40376,54.595406
73.56321,53.199406
95.13926,69.66168
103.97244,66.49064
65.96929,47.48809
81.25386,57.820103
86.36154,57.31073
92.42379,67.19231
68.11478,47.946205
71.20676,49.8056
108.45938,66.43607
93.277695,68.67151
67.05038,47.676136
79.78789,57.88972
74.87908,55.29064
79.49469,57.880463
89.37785,58.54345
106.95904,65.77898
80.78044,57.8599
98.47919,68.99922
84.019806,57.455715
73.8353,53.665348
80.66119,57.867683
76.77666,57.116966
74.26545,54.377274
69.432274,48.48101
63.283775,47.121803
63.616543,47.172565
72.93113,52.12703
81.7729,57.7632
71.50804,50.129574
93.61654,69.01099
90.27472,59.875393
84.668304,57.37953
65.262024,47.390335
101.74204,67.508316
99.37213,68.62584
73.00548,52.24997
92.74753,67.887184
74.56288,54.83807
108.43132,66.40654
76.15849,56.679646
73.49158,53.07614
108.5905,66.59001
93.21804,68.60042
90.96892,61.76042
88.533585,57.875877
106.84482,65.77513
75.996254,56.541702
108.8618,67.01758
94.893555,69.6239
109.11921,67.65021
62.290203,46.9384
108.11298,66.13849
68.867386,48.21734
92.12111,66.33725
68.91644,48.237976
96.1428,69.64946
81.79806,57.760162
90.74434,61.036118
71.67781,50.327328
85.39762,57.320107
95.05159,69.650406
89.69113,58.91278
86.577896,57.323666
108.88264,67.05848
90.68565,60.866447
97.15971,69.44433
66.14101,47.514088
108.9283,67.15332
65.798256,47.46324
78.87851,57.82636
60.722965,46.49486
89.374695,58.540127
64.84805,47.33703
62.297134,46.939888
72.50768,51.456062
75.22398,55.73528
92.5421,67.46969
91.76163,65.02105
69.87601,48.73332
90.026215,59.413406
91.29399,63.01921
67.267746,47.722847
66.251755,47.531487
88.24255,57.725243
62.820454,47.043148
62.73942,47.02825
68.796814,48.188347
92.12008,66.33395
64.82164,47.33366
59.841995,46.123955
77.25728,57.37197
79.44801,57.878082
76.39547,56.863194
87.33865,57.429527
62.4294,46.967644
91.37279,63.354008
91.049194,62.049133
90.173035,59.67565
73.6677,53.379017
77.249405,57.368324
80.15135,57.888752
83.94881,57.46496
108.97946,67.269
90.77887,61.139618
96.19858,69.64244
108.233795,66.22717
107.11398,65.79105
95.94585,69.66956
69.41282,48.47092
103.511894,66.6853
90.42824,60.208614
95.237015,69.67156
99.81618,68.426125
70.455025,49.13498
101.509,67.62091
63.73288,47.189407
92.998146,68.30362
80.008644,57.890663
86.428856,57.31406
100.82977,67.948814
96.847244,69.52318
74.652145,54.969982
61.066086,46.612896
92.36841,67.052155
88.572655,57.898685
77.57456,57.50553
70.86509,49.477745
73.710464,53.452335
91.25897,62.87323
102.26785,67.256035
86.29464,57.30802
66.691666,47.606483
67.401405,47.75348
89.673546,58.889763
75.80003,56.360825
89.972694,59.32484
107.98245,66.0575
60.397987,46.370056
90.2498,59.824986
76.785515,57.122295
103.911285,66.51577
74.90802,55.329956
92.933075,68.20419
70.27634,49.001396
87.874054,57.57702
92.40972,67.15736
98.88393,68.835434
78.39067,57.743687
102.309204,67.23637
79.44081,57.87769
85.28304,57.327034
60.90506,46.55918
70.797195,49.41732
73.06592,52.350796
67.66361,47.81824
77.43954,57.45182
94.51079,69.52278
106.62866,65.7784
88.09761,57.66174
89.39226,58.558678
61.616455,46.775974
106.319016,65.804054
62.7649,47.032974
64.595535,47.3048
104.239914,66.38359
99.524216,68.55829
90.66366,60.804863
72.28844,51.132133
94.44414,69.499176
68.64038,48.1269
92.69243,67.78257
94.68503,69.57575
70.26659,48.994358
104.274055,66.370285
79.4138,57.876167
81.50524,57.79402
93.096344,68.44329
77.229965,57.35925
65.51477,47.42392
69.378174,48.45314
87.4516,57.454853
81.048904,57.83896
88.20528,57.70822
83.776,57.488
83.94396,57.46559
65.93984,47.48374
70.21596,48.95827
82.9376,57.606106
89.321144,58.485
62.5077,46.98352
73.66999,53.382942
64.68513,47.31625
87.32826,57.427334
65.08573,47.367462
91.49481,63.88324
75.22975,55.742268
71.975784,50.701736
76.896385,57.1869
102.75618,67.02618
74.49072,54.72918
85.680405,57.30765
108.2257,66.22078
75.2616,55.78057
96.17354,69.64566
75.10692,55.590275
86.42436,57.31382
82.18588,57.71076
85.36704,57.321854
92.37315,67.064415
97.39812,69.37628
86.3952,57.31232
88.03941,57.6382
81.61376,57.781876
109.1712,67.82467
94.671165,69.571976
105.836685,65.88438
60.13468,46.258873
106.91206,65.7769
78.7502,57.80835
90.17346,59.676453
75.03909,55.503464
63.017246,47.07784
87.971695,57.612152
83.36291,57.545387
76.465996,56.913887
98.47157,69.0022
63.293552,47.123356
91.36345,63.313972
67.70365,47.828712
108.97833,67.26634
76.536995,56.963158
89.993866,59.35945
90.68078,60.852715
84.48042,57.39982
104.75719,66.192154
75.93179,56.48401
74.25791,54.3652
95.47139,69.68443
66.915474,47.648926
64.950615,47.35013
99.93589,68.37108
69.47873,48.505417
70.85028,49.464436
107.775665,65.95579
93.13153,68.49046
96.121155,69.65203
87.57619,57.48619
90.154015,59.639984
108.58771,66.58643
92.21913,66.63875
91.19661,62.61863
99.95255,68.36338
60.347546,46.349476
89.694115,58.916718
64.39439,47.278908
63.84976,47.205925
86.998276,57.36927
61.12748,46.632626
89.4724,58.646294
85.695496,57.30719
71.49108,50.110435
91.66434,64.61559
94.34387,69.45993
97.142685,69.44894
87.99004,57.61907
62.495224,46.981018
85.49915,57.314842
80.33759,57.883514
92.62985,67.657394
70.42611,49.112743
84.324425,57.417862
83.71477,57.496323
59.839375,46.122692
100.65938,68.03047
83.208374,57.56737
100.0634,68.31197
64.75923,47.325703
88.327545,57.7659
106.79958,65.77472
91.09119,62.20636
77.039,57.26452
108.69649,66.736824
60.147415,46.26447
70.53592,49.198494
109.16109,67.78894
67.29219,47.728336
77.498215,57.47571
69.952415,48.781338
109.23152,68.058586
97.67729,69.28889
61.88685,46.845615
88.07303,57.651665
91.84337,65.347855
97.819046,69.24161
60.637165,46.46319
90.533325,60.46108
102.37935,67.20309
"""

data = np.loadtxt(StringIO(data_str.strip()), delimiter=",")
x = data[:,0].astype(float)
y = data[:,1].astype(float)
N = len(x)
print(f"Loaded {N} points.")

def project_t_s(theta, X, xpts=x, ypts=y):
    c = np.cos(theta); s = np.sin(theta)
    vx = xpts - X
    vy = ypts - 42.0
    t = vx*c + vy*s
    sdat = -vx*s + vy*c
    return t, sdat

def s_model(M, t):
    return np.exp(M*t) * np.sin(0.3*t)

def inner_best_M(theta, X):
    t, sdat = project_t_s(theta, X)

    if np.any(t < 6) or np.any(t > 60):
        return {"M":0.0, "mse":1e8, "success":False}
    if SCIPY:
        def mse_M(M):
            pred = s_model(M, t)
            return np.mean((sdat - pred)**2)
        res = minimize_scalar(mse_M, bounds=(-0.05,0.05), method="bounded", options={"xatol":1e-8})
        return {"M":float(res.x), "mse":float(res.fun), "success":res.success}
    else:
        Ms = np.linspace(-0.05,0.05,401)
        errs = [np.mean((sdat - s_model(M, t))**2) for M in Ms]
        idx = int(np.argmin(errs))
        return {"M":float(Ms[idx]), "mse":float(errs[idx]), "success":True}

def outer_objective_theta_X(vars):
    theta, X = float(vars[0]), float(vars[1])
    if not (0.0 <= theta <= np.deg2rad(50.0) and 0.0 <= X <= 100.0):
        return 1e9
    inner = inner_best_M(theta, X)
    return inner["mse"]

# initial guess using PCA on (x, y-42)
xy = np.column_stack([x, y-42.0])
C = np.cov(xy.T)
eigvals, eigvecs = np.linalg.eig(C)
pc1 = eigvecs[:, np.argmax(eigvals)]
theta0 = np.arctan2(pc1[1], pc1[0])
theta0 = float(np.clip(theta0, 0.0, np.deg2rad(50.0)))
c0, s0 = np.cos(theta0), np.sin(theta0)
mean_vy = np.mean(y-42.0)
mean_x = np.mean(x)
t_target = 33.0
if abs(c0) < 1e-6:
    X0 = min(max(mean_x, 0.0), 100.0)
else:
    X0 = mean_x - (t_target - mean_vy*s0)/c0
X0 = float(np.clip(X0, 0.0, 100.0))
print(f"Initial guess: theta={degrees(theta0):.6f} deg, X={X0:.6f}")

if SCIPY:
    x0 = np.array([theta0, X0])
    bounds = [(0.0, np.deg2rad(50.0)), (0.0, 100.0)]
    res_outer = minimize(lambda v: outer_objective_theta_X(v), x0, method="L-BFGS-B", bounds=bounds,
                         options={"ftol":1e-12, "maxiter":500})
    theta_opt, X_opt = float(res_outer.x[0]), float(res_outer.x[1])
    outer_success = res_outer.success
    outer_fval = float(res_outer.fun)
else:
    th_grid = np.linspace(0, np.deg2rad(50), 101)
    X_grid = np.linspace(0,100,201)
    best = {"mse":1e9}
    for th in th_grid:
        for Xcand in X_grid:
            val = outer_objective_theta_X((th,Xcand))
            if val < best["mse"]:
                best = {"mse":val, "theta":th, "X":Xcand}
    theta_opt, X_opt, outer_fval = best["theta"], best["X"], best["mse"]
    outer_success = True

inner = inner_best_M(theta_opt, X_opt)
M_opt = inner["M"]


t_fit, sdat = project_t_s(theta_opt, X_opt)
x_pred = X_opt + t_fit*np.cos(theta_opt) - np.exp(M_opt*t_fit)*np.sin(0.3*t_fit)*np.sin(theta_opt)
y_pred = 42.0 + t_fit*np.sin(theta_opt) + np.exp(M_opt*t_fit)*np.sin(0.3*t_fit)*np.cos(theta_opt)


idx_sort = np.argsort(t_fit)
x_obs_sorted = x[idx_sort]; y_obs_sorted = y[idx_sort]
K = N
t_uniform = np.linspace(6.0, 60.0, K)
x_uniform_pred = X_opt + t_uniform*np.cos(theta_opt) - np.exp(M_opt*t_uniform)*np.sin(0.3*t_uniform)*np.sin(theta_opt)
y_uniform_pred = 42.0 + t_uniform*np.sin(theta_opt) + np.exp(M_opt*t_uniform)*np.sin(0.3*t_uniform)*np.cos(theta_opt)
abs_diff = np.abs(x_obs_sorted - x_uniform_pred) + np.abs(y_obs_sorted - y_uniform_pred)
mean_L1 = np.mean(abs_diff)
mean_L2 = np.mean(np.sqrt((x_obs_sorted - x_uniform_pred)**2 + (y_obs_sorted - y_uniform_pred)**2))
residuals_s = sdat - s_model(M_opt, t_fit)
mse_s = np.mean(residuals_s**2)


print("---- Fit results ----")
print(f"theta (rad) = {theta_opt:.8f}, theta (deg) = {degrees(theta_opt):.6f}")
print(f"X = {X_opt:.6f}")
print(f"M = {M_opt:.8f}")
print(f"final mse (perp s) = {mse_s:.6e}")
print(f"outer objective value = {outer_fval:.6e}, outer_success={outer_success}")
print(f"mean L1 (|dx|+|dy|) = {mean_L1:.6f}")
print(f"mean Euclidean (L2) = {mean_L2:.6f}")


latex_rad = r"\left(t\cos({th:.6f})-e^{{{M:.6f}|t|}}路\sin(0.3t)\sin({th:.6f})+{X:.6f},\ 42+t\sin({th:.6f})+e^{{{M:.6f}|t|}}路\sin(0.3t)\cos({th:.6f})\right)".format(
    th=theta_opt, M=M_opt, X=X_opt)
latex_deg = r"\left(t\cos({thdeg:.4f}^\circ)-e^{{{M:.6f}|t|}}路\sin(0.3t)\sin({thdeg:.4f}^\circ)+{X:.6f},\ 42+t\sin({thdeg:.4f}^\circ)+e^{{{M:.6f}|t|}}路\sin(0.3t)\cos({thdeg:.4f}^\circ)\right)".format(
    thdeg=degrees(theta_opt), M=M_opt, X=X_opt)
print("\nLaTeX (rad):\n", latex_rad)
print("\nLaTeX (deg):\n", latex_deg)

# Save summary to CSV
import pandas as pd
summary = pd.DataFrame([{
    "theta_deg": degrees(theta_opt),
    "theta_rad": theta_opt,
    "M": M_opt,
    "X": X_opt,
    "mean_L1": mean_L1,
    "mean_L2": mean_L2,
    "mse_s": mse_s
}])
summary.to_csv("fit_summary.csv", index=False)
print("Saved fit_summary.csv")


try:
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8,6))
    plt.scatter(x, y, s=6, alpha=0.4, label='observed')
    plt.scatter(x_pred, y_pred, s=6, alpha=0.4, label='predicted (matched t)', marker='x')
    # predicted uniform curve (smooth)
    tt = np.linspace(6,60,500)
    xp = X_opt + tt*np.cos(theta_opt) - np.exp(M_opt*tt)*np.sin(0.3*tt)*np.sin(theta_opt)
    yp = 42.0 + tt*np.sin(theta_opt) + np.exp(M_opt*tt)*np.sin(0.3*tt)*np.cos(theta_opt)
    plt.plot(xp, yp, 'r-', lw=1.0, label='predicted (uniform t)')
    plt.legend()
    plt.xlabel('x'); plt.ylabel('y'); plt.title('Observed vs Predicted')
    plt.axis('equal'); plt.grid(True)
    plt.show()
except Exception:
    pass
```
